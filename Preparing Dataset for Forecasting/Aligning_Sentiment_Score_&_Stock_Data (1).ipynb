```python
import pandas as pd
import numpy as np
```

```python
stock_df= pd.read_excel(io='/content/Prudential_new.xlsx', sheet_name='Sheet3')
```

```python
stock_df
```

```python
date_range = pd.date_range(start='2017-01-01', end='2023-05-31')

# Create a DataFrame with the complete date range
complete_dates_df = pd.DataFrame(date_range, columns=['Date'])

# Merge the stock data with the complete date range
stock_df = pd.merge(complete_dates_df, stock_df, how='left', on='Date')
```

```python
stock_df
```

```python
sentiment_daily=pd.read_excel('/content/stock_tweets_sentiment(VADER+BERT).xlsx')
```

```python
sentiment_daily
```

```python
sentiment_daily['Date'] = pd.to_datetime(sentiment_daily['Date'])
stock_df['Date'] = pd.to_datetime(stock_df['Date'])

# Merge the dataframes, only including the 'combined_score' column from sentiment_daily
merged_df = pd.merge(stock_df, sentiment_daily[['Date','negative','neutral', 'positive', 'combined_score']], how='left', left_on='Date', right_on='Date')
```

```python
merged_df
```

```python
df_cleaned = merged_df.dropna(subset=['Close', 'combined_score'], how='all')
df_cleaned = df_cleaned.dropna(subset=['negative', 'neutral', 'positive'], how='all')
```

```python
df_cleaned
```

```python
df_cleaned['Close'] = df_cleaned['Close'].fillna(method='ffill')
```

```python
df_cleaned
```

```python
df_cleaned['Close'] = df_cleaned['Close'].fillna(method='bfill')
```

```python
df_cleaned.to_excel('Stock_Forecasting_data(VADER+BERT).xlsx', index=False)
```
