```python
import pandas as pd
import pandas_datareader.data as web
from datetime import datetime

start_date = datetime(2017, 1, 1)
end_date = datetime.today()

gdp_data = web.DataReader('UKNGDP', 'fred', start_date, end_date)
```

```python
gdp_data['Date'] = pd.to_datetime(gdp_data.index)
```

```python
# Drop the index and reset it
gdp_data.reset_index(drop=True, inplace=True)

# Reorder the columns
gdp_data = gdp_data[['Date', 'UKNGDP']]
```

```python
gdp_data = pd.DataFrame(gdp_data)
gdp_data['Date'] = pd.to_datetime(gdp_data['Date'])

# Forward-fill GDP values for each day in the sentiment/stock dataset
gdp_data = gdp_data.set_index('Date').resample('D').ffill()

# Merge the datasets
df_combined = pd.merge(df, gdp_data, left_on='Date', right_index=True, how='left')

# Fill NaN values in 'Close' column with the previous value
df_combined['Close'] = df_combined['Close'].fillna(method='ffill')
```

```python
moneysupply = pd.DataFrame(moneysupply)
moneysupply['Date'] = pd.to_datetime(moneysupply['Date'])

# Forward-fill GDP values for each day in the sentiment/stock dataset
moneysupply = moneysupply.set_index('Date').resample('D').ffill()

# Merge the datasets
df_final = pd.merge(df_combined, moneysupply, left_on='Date', right_index=True, how='left')

# Fill NaN values in 'Close' column with the previous value
df_final['Close'] = df_final['Close'].fillna(method='ffill')
```

```python
df_final['Brexit'] = (df_final['Date'] > '2020-01-31').astype(int)
```

```python
df_final['Date'] = df_final['Date'].dt.strftime('%Y-%m-%d')
```

```python
pm_terms = [
    ("Theresa May", "2016-07-13", "2019-07-24"),
    ("Boris Johnson", "2019-07-24", "2022-09-06"),
    ("Liz Truss", "2022-09-06", "2022-10-25"),
    ("Rishi Sunak", "2022-10-25", None)
]
```

```python
def get_pm_code(date_str):
    date = datetime.strptime(date_str, "%Y-%m-%d")
    for i, (pm, start, end) in enumerate(pm_terms, start=1):
        start_date = datetime.strptime(start, "%Y-%m-%d")
        end_date = datetime.strptime(end, "%Y-%m-%d") if end else datetime.now()
        if start_date <= date <= end_date:
            return i
    return None
```

```python
df_final['PM_Code'] = df_final['Date'].apply(get_pm_code)
```
